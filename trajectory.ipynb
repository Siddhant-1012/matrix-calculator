```python
import matplotlib.pyplot as plt
import numpy as np

# --- Simulation Parameters ---
max_time = 10 # Maximum time for simulation
initial_speed = 30 # Initial launch speed (m/s)
gravity = 9.8    # Acceleration due to gravity (m/s^2)
launch_angles = [30,45,60] # Launch angles in degrees

# --- Plotting Setup ---
'''plt.figure(figsize=(10, 6))''' # Set figure size for better visualization

# --- Simulate and Plot Trajectories ---
for angle_deg in launch_angles:
    # Convert angle from degrees to radians
    rad = np.radians(angle_deg)

    # Generate time points for the trajectory
    # 501 points for smooth curve from 0 to max_time
    t = np.linspace(0, max_time, 501)

    # Calculate x (horizontal) and y (vertical) positions
    x = initial_speed * np.cos(rad) * t
    y = initial_speed * np.sin(rad) * t - 0.5 * gravity * t**2

    # Filter out points where y (height) is negative (object has hit the ground)
    mask = y >= 0
    x = x[mask]
    y = y[mask]


    # Plot the trajectory
    plt.plot(x,y, label=f'{angle_deg}°')

# --- Customize Plot ---
plt.title('Projectile Motion at Different Launch Angles') # Add a title
plt.xlabel('Horizontal Distance (m)') # Add x-axis label
plt.ylabel('Vertical Distance (m)')   # Add y-axis label
plt.legend(title='Angle') # Add a legend to distinguish trajectories
plt.grid(True) # Add a grid for better readability
plt.ylim(bottom=0) # Ensure y-axis starts at 0

# Display the plot
plt.show()
```
