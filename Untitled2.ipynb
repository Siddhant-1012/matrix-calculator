```python
import numpy as np

import numpy as np

A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("Addition")
print(A + B)

print("\nSubtraction")
print(A - B)

print("\nMultiplication")
print(A @ B)

print("\nDeterminant of A")
print(np.linalg.det(A))

print("\nInverse of A")
print(np.linalg.inv(A))

print("\nTranspose of A")
print(A.T)

```
