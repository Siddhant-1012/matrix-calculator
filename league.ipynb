```python
import random
teams = {"arsenal": 0.7, "man city": 0.6, "man utd": 0.5, "aston villa": 0.4}
Simulations=100000
HOME_ADV = 0.1
DRAW_PROB = 0.3 # Renamed to avoid confusion with existing 'DRAW' logic
win_count={"arsenal": 0, "man city": 0, "man utd": 0, "aston villa": 0}
'''print(f"{'Home Team':<15} {'Away Team':<15} {'P(Home Win)':<15} {'P(Draw)':<15} {'P(Away Win)':<15} {'Random':<10} {'Winner':<15}")
print("-"*100)'''
for _ in range (Simulations):
  points={"arsenal": 0, "man city": 0, "man utd": 0, "aston villa": 0}
  for team_home in teams:
      for team_away in teams:
          if team_home != team_away:
              # Calculate base strengths including home advantage
              strength_home = teams[team_home] + HOME_ADV
              strength_away = teams[team_away]
              total_strength_base = strength_home + strength_away

              # Calculate conditional probabilities for home/away win if no draw
              prob_home_win_conditional = strength_home / total_strength_base
              prob_away_win_conditional = strength_away / total_strength_base

              # Distribute probabilities for three outcomes (Home Win, Draw, Away Win)
              # The DRAW_PROB is the fixed probability of a draw
              # The remaining probability (1 - DRAW_PROB) is split proportionally between home and away wins
              prob_home_win = prob_home_win_conditional * (1 - DRAW_PROB)
              prob_draw = DRAW_PROB
              prob_away_win = prob_away_win_conditional * (1 - DRAW_PROB)

              # Ensure probabilities sum to 1 (due to floating point, might be slightly off)
              # total_prob = prob_home_win + prob_draw + prob_away_win

              r = random.random()
              winner = ""

              if r < prob_home_win:
                  winner = team_home
                  points[winner]=points[winner]+3
              elif r < prob_home_win + prob_draw:
                  winner = 'Draw'
                  points[team_home]=points[team_home]+1
                  points[team_away]=points[team_away]+1
              else:
                  winner = team_away
                  points[winner]=points[winner]+3

              '''print(f"{team_home:<15} {team_away:<15} {prob_home_win:<15.2f} {prob_draw:<15.2f} {prob_away_win:<15.2f} {r:<10.2f} {winner:<15}")'''
  x = max(points, key=points.get)
  win_count[x]=win_count[x]+1
print(win_count)
for team in win_count:
  print(team,f'{win_count[team]/Simulations:.2f}')
```
