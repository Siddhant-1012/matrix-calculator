```python
groups = {
    "A": ["Mexico", "South Africa", "South Korea", "Czechia"],
    "B": ["Canada", "Bosnia and Herzegovina", "Qatar", "Switzerland"],
    "C": ["Brazil", "Morocco", "Haiti", "Scotland"],
    "D": ["United States", "Paraguay", "Australia", "Turkey"],
    "E": ["Germany", "Curacao", "Ivory Coast", "Ecuador"],
    "F": ["Netherlands", "Japan", "Sweden", "Tunisia"],
    "G": ["Belgium", "Egypt", "Iran", "New Zealand"],
    "H": ["Spain", "Cape Verde", "Saudi Arabia", "Uruguay"],
    "I": ["France", "Senegal", "Iraq", "Norway"],
    "J": ["Argentina", "Algeria", "Austria", "Jordan"],
    "K": ["Portugal", "DR Congo", "Uzbekistan", "Colombia"],
    "L": ["England", "Croatia", "Ghana", "Panama"],
}
teams = [team for group in groups.values() for team in group]
team_strength = {
    "Mexico": 78,
    "South Africa": 60,
    "South Korea": 75,
    "Czechia": 72,
    "Canada": 76,
    "Bosnia and Herzegovina": 68,
    "Qatar": 62,
    "Switzerland": 82,
    "Brazil": 95,
    "Morocco": 84,
    "Haiti": 55,
    "Scotland": 78,
    "United States": 84,
    "Paraguay": 74,
    "Australia": 76,
    "Turkey": 77,
    "Germany": 92,
    "Curacao": 52,
    "Ivory Coast": 78,
    "Ecuador": 80,
    "Netherlands": 90,
    "Japan": 85,
    "Sweden": 79,
    "Tunisia": 70,
    "Belgium": 86,
    "Egypt": 73,
    "Iran": 77,
    "New Zealand": 66,
    "Spain": 94,
    "Cape Verde": 67,
    "Saudi Arabia": 71,
    "Uruguay": 87,
    "France": 96,
    "Senegal": 83,
    "Iraq": 64,
    "Norway": 81,
    "Argentina": 97,
    "Algeria": 76,
    "Austria": 82,
    "Jordan": 61,
    "Portugal": 91,
    "DR Congo": 72,
    "Uzbekistan": 74,
    "Colombia": 86,
    "England": 93,
    "Croatia": 84,
    "Ghana": 75,
    "Panama": 67,
}
score_table=[]
point_table = {team: {'pt':0, 'gs':0, 'gc':0, 'gd':0, 'group':''} for team in teams}
```

```python
from itertools import combinations

def get_score_table(group_name):
    score_table_group = [] # Use a local variable to collect scores for the current group
    teams = groups[group_name]
    fixture = list(combinations(teams, 2))
    for x in fixture:
        score = simulate_match(team_strength[x[0]], team_strength[x[1]])
        score_table_group.append((x[0], x[1], score[0], score[1], group_name))
    return score_table_group

score_table = [] # Clear and re-populate the global score_table
for group_name_key in groups.keys():
    score_table.extend(get_score_table(group_name_key))

# Optional: print all scores collected
print(*score_table, sep='\n')
```

```python
import pandas as pd # Import pandas for tabular display

def get_point_table(score_table):
    global point_table # Declare intent to modify the global point_table
    # Re-initialize point_table to ensure a fresh calculation for all teams
    point_table = {team: {'pt':0, 'gs':0, 'gc':0, 'gd':0, 'group':''} for team in teams}

    for score in score_table:
        team1, team2, goals1, goals2, group_name = score

        # Update team1's stats
        point_table[team1]['gs'] += goals1
        point_table[team1]['gc'] += goals2
        point_table[team1]['gd'] += goals1 - goals2
        point_table[team1]['group'] = group_name

        # Update team2's stats
        point_table[team2]['gs'] += goals2
        point_table[team2]['gc'] += goals1
        point_table[team2]['gd'] += goals2 - goals1
        point_table[team2]['group'] = group_name

        if goals1 > goals2:
            point_table[team1]['pt'] += 3
        elif goals1 == goals2:
            point_table[team1]['pt'] += 1
            point_table[team2]['pt'] += 1
        else:
            point_table[team2]['pt'] += 3

# Call get_point_table with the globally populated score_table
get_point_table(score_table)

# Convert the point_table dictionary to a pandas DataFrame for tabular display
df_point_table = pd.DataFrame.from_dict(point_table, orient='index')
df_point_table.index.name = 'Team'

# Sort the DataFrame for better readability: by group, then points, goal difference, and goals scored
df_point_table = df_point_table.sort_values(by=['group', 'pt', 'gd', 'gs'], ascending=[True, False, False, False])
df_point_table['position'] = df_point_table.groupby('group').cumcount() + 1
df_team12=df_point_table[df_point_table['position'] <3]
df_team3=df_point_table[df_point_table['position'] ==3].sort_values(by=[ 'pt', 'gd', 'gs'], ascending=[ False, False, False]).nlargest(8, 'pt')
qualified_teams= pd.concat([df_team12,df_team3])
display(df_team12)
display(df_team3)
#display( df_point_table[df_point_table['position'] <3])
#display( df_point_table[df_point_table['position'] ==3].sort_values(by=[ 'pt', 'gd', 'gs'], ascending=[ False, False, False]).nlargest(8, 'pt'))
display(qualified_teams)

# Display the DataFramedisplay(df_point_table)
```

```python
import random

def simulate_match(strength1, strength2):

    diff = strength1 - strength2

    goals1 = max(
        0,
        round(1.5 + diff / 25 + random.uniform(-1, 1))
    )

    goals2 = max(
        0,
        round(1.5 - diff / 25 + random.uniform(-1, 1))
    )

    return goals1, goals2
```

```python
qualified_teams_all_groups = [] # Initialize list to store qualified teams

for group_name in groups.keys():
    # Filter df_point_table for the current group and sort by points, GD, GS
    group_df = df_point_table[df_point_table['group'] == group_name].sort_values(
        by=['pt', 'gd', 'gs'], ascending=[False, False, False]
    )

    print(f"Group {group_name}:")
    print(group_df[['pt', 'gd', 'gs']])

    # Get the top two teams from the sorted group DataFrame
    top_teams = group_df.index[:2].tolist()
    qualified_teams_all_groups.extend(top_teams)

print("\n--- Qualified Teams (Top 2 from each group) for Round of 32 ---")
print(qualified_teams_all_groups)
```
