```python
from itertools import combinations

def get_score_table(group_name):
    score_table = []
    teams = world_cup_2026[group_name]
    fixture = list(combinations(teams.keys(), 2))
    for x in fixture:
        #print(x)
        score = simulate_match(teams[x[0]], teams[x[1]])
        score_table.append((x[0], x[1], score[0], score[1]))
    return score_table
```

```python
def get_point_table(score_table, group_name):
    point_table = dict.fromkeys(world_cup_2026[group_name], 0)

    for score in score_table:
        if score[2] > score[3]:
            point_table[score[0]] = point_table[score[0]] + 3
        elif score[2] == score[3]:
            point_table[score[0]] = point_table[score[0]] + 1
            point_table[score[1]] = point_table[score[1]] + 1
        else:
            point_table[score[1]] = point_table[score[1]] + 3
    return point_table
```

```python
import random

def simulate_match(strength1, strength2):

    diff = strength1 - strength2

    goals1 = max(
        0,
        round(1.5 + diff / 25 + random.uniform(-1, 1))
    )

    goals2 = max(
        0,
        round(1.5 - diff / 25 + random.uniform(-1, 1))
    )

    return goals1, goals2
```

```python
for group_name in list(world_cup_2026.keys()):
    score_table = get_score_table(group_name)

    points_table = get_point_table(score_table, group_name)



    # Sort by value descending
    sorted_dict_desc = dict(sorted(points_table.items(), key=lambda item: item[1], reverse=True))

    print(group_name, sorted_dict_desc)
```
